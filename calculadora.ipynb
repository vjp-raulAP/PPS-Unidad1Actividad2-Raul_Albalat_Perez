```python
# Jupyter Notebook: Creación de un módulo en Python

## Paso 1: Creación del módulo con funciones básicas

# Definimos las funciones directamente en el notebook

def suma(a, b):
    """Devuelve la suma de dos números."""
    return a + b

def resta(a, b):
    """Devuelve la resta de dos números."""
    return a - b

def multiplicacion(a, b):
    """Devuelve la multiplicación de dos números."""
    return a * b

def division(a, b):
    """Devuelve la división de dos números. Si b es 0, lanza una excepción."""
    if b == 0:
        raise ValueError("No se puede dividir entre cero.")
    return a / b

def isNumber(value):
    """Devuelve True si el valor es un número, False en caso contrario."""
    return isinstance(value, (int, float))

def mayorCero(value):
    """Devuelve True si el número es mayor que cero, False en caso contrario."""
    if not isNumber(value):
        raise ValueError("El valor no es un número.")
    return value > 0

## Paso 2: Creación del programa que utiliza las funciones

# Creamos un programa que solicita dos números al usuario y permite seleccionar una operación para ejecutarla.

def main():
    """Programa principal que realiza operaciones matemáticas básicas usando las funciones definidas."""
    print("Bienvenido al programa de operaciones matemáticas")

    try:
        # Solicitar al usuario los números
        num1 = float(input("Introduce el primer número: "))
        num2 = float(input("Introduce el segundo número: "))

        # Mostrar menú de opciones
        print("\nSeleccione una operación:")
        print("1. Suma")
        print("2. Resta")
        print("3. Multiplicación")
        print("4. División")

        opcion = input("Introduce el número de la operación que deseas realizar: ")

        # Ejecutar la operación seleccionada
        if opcion == "1":
            print(f"Resultado de la suma: {suma(num1, num2)}")
        elif opcion == "2":
            print(f"Resultado de la resta: {resta(num1, num2)}")
        elif opcion == "3":
            print(f"Resultado de la multiplicación: {multiplicacion(num1, num2)}")
        elif opcion == "4":
            try:
                print(f"Resultado de la división: {division(num1, num2)}")
            except ValueError as e:
                print(e)
        else:
            print("Opción no válida. Por favor, selecciona una opción entre 1 y 4.")

    except ValueError:
        print("Por favor, introduce valores numéricos válidos.")

# Ejecutar el programa principal en el entorno del notebook
main()
```
